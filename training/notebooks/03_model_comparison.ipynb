# 03 · Model comparison

Rank trained versions and identify the production candidate.

```
import json, sys
from pathlib import Path
ROOT = Path.cwd()
while ROOT != ROOT.parent and not (ROOT / 'weights').is_dir():
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / 'training'))
sys.path.insert(0, str(ROOT))
from evaluate import compare_models
VERSIONS = ['v1_baseline', 'v2_improved', 'v3_final']
```

```
metrics = {}
for v in VERSIONS:
    mp = ROOT / 'weights' / v / 'metrics.json'
    if mp.is_file():
        metrics[v] = json.loads(mp.read_text()).get('metrics', {})
if not metrics:
    print('No trained versions yet. Train v1/v2/v3 and record metrics first.')
else:
    print('Loaded metrics for:', list(metrics))
    print('Comparison:', compare_models(metrics))
```
