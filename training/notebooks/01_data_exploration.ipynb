# 01 · Data exploration

Summary statistics for the prepared specimen images.

```
import json
from pathlib import Path
import matplotlib.pyplot as plt

TEAL, ABYSS, HAIR = '#2BE1CB', '#0B1620', '#20404D'
ROOT = Path.cwd()
while ROOT != ROOT.parent and not (ROOT / 'data').is_dir():
    ROOT = ROOT.parent
MANIFEST = ROOT / 'data' / 'processed' / 'image_manifest.json'
print('manifest:', MANIFEST, '| exists:', MANIFEST.is_file())
```

```
records = json.loads(MANIFEST.read_text()) if MANIFEST.is_file() else []
if not records:
    print('No manifest yet. Run scripts/prepare_dataset.py after importing raw images.')
else:
    widths = [r['width'] for r in records]
    heights = [r['height'] for r in records]
    print(f'Images: {len(records)}')
    print(f'Width  min/mean/max: {min(widths)}/{sum(widths)/len(widths):.0f}/{max(widths)}')
    print(f'Height min/mean/max: {min(heights)}/{sum(heights)/len(heights):.0f}/{max(heights)}')
```

```
if records:
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.scatter([r['width'] for r in records], [r['height'] for r in records],
               c=TEAL, s=12)
    ax.set_facecolor(ABYSS); fig.set_facecolor(ABYSS)
    ax.set_xlabel('width (px)', color='#EAF4F2'); ax.set_ylabel('height (px)', color='#EAF4F2')
    ax.tick_params(colors='#EAF4F2'); ax.grid(True, color=HAIR, linewidth=0.5)
    plt.show()
else:
    print('Nothing to plot yet.')
```
