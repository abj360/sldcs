# 02 · Training analysis

Loss and mAP curves for a completed training run.

```
import csv
from pathlib import Path
import matplotlib.pyplot as plt

TEAL, ABYSS, HAIR = '#2BE1CB', '#0B1620', '#20404D'
ROOT = Path.cwd()
while ROOT != ROOT.parent and not (ROOT / 'training').is_dir():
    ROOT = ROOT.parent
# Point this at the run you want to analyze.
RESULTS = ROOT / 'runs' / 'train' / 'v1_baseline' / 'results.csv'
print('results.csv:', RESULTS, '| exists:', RESULTS.is_file())
```

```
rows = []
if RESULTS.is_file():
    with RESULTS.open() as fh:
        rows = [{k.strip(): v.strip() for k, v in r.items()} for r in csv.DictReader(fh)]
if not rows:
    print('No training results yet. Run training/train.py first.')
else:
    print(f'{len(rows)} epochs recorded. Columns:', list(rows[0]))
```

```
if rows:
    epochs = [int(float(r['epoch'])) for r in rows]
    m = [float(r.get('metrics/mAP_0.5', 0) or 0) for r in rows]
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.plot(epochs, m, color=TEAL, linewidth=2)
    ax.set_facecolor(ABYSS); fig.set_facecolor(ABYSS)
    ax.set_xlabel('epoch', color='#EAF4F2'); ax.set_ylabel('mAP@0.5', color='#EAF4F2')
    ax.tick_params(colors='#EAF4F2'); ax.grid(True, color=HAIR, linewidth=0.5)
    plt.show()
else:
    print('Nothing to plot yet.')
```
